```python
import pandas as pd

# monthly rate
rf = pd.read_csv('./F-F_Research_Data_Factors.csv', index_col = 0) 
rf.index = pd.to_datetime(rf.index, format = '%Y%m')
rf.index = rf.index.to_period('M').to_timestamp('M')
rf
```

```python
# 1990-04-30 ~ 2021-10-31 슬라이싱
rf = rf.loc['1990-04-30':'2021-10-31']

# RF만 저장
rf = rf['RF']
rf = pd.DataFrame(rf)
rf
```

```python
# % 단위로 변경
rf['RF(%)'] = rf['RF'] * 100
```

```python
rf.to_csv('rf.csv', index=False)
```
